**Рассчет мощности двигателя автомобиля BMV e36 1993 года выпуска на преодоление аэродинамического сопротивления в зависимости от скорости.**

```
import matplotlib.pyplot as plt 
import seaborn as sns
import torch as tr
```

```
s = 1.8 
v = tr.linspace(0,200, 1000)
k = 0.32
p = 1.2 
P = p * (v * 1000 / 60 / 60)**3 * k *s / (2 * 700)

sns.scatterplot(v, P)
plt.title('Мощность BMW e36 на преодоление аэродинамического сопротивления')
plt.gcf().set_size_inches(10,7)
plt.xlabel('Скорость [км/ч]')
plt.ylabel('Мощность [л.с.]')
plt.grid()
```
